```python
# notebooks/EDA.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Load the exported processed dataset
df = pd.read_csv("../data/export_nhl_players.csv")

#  Confirm column names
print("Columns in data:", df.columns.tolist())

# ========== PLOT 1: Value Score Distribution ==========
plt.figure(figsize=(8, 5))
sns.histplot(df["ValueScore"], bins=10, kde=True)
plt.title("Distribution of Player Value Scores")
plt.xlabel("ValueScore")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("../outputs/graphs/value_score_distribution.png")
plt.show()

# ========== PLOT 2: Cap Hit vs Value Score ==========
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="CapHit", y="ValueScore", hue="Team")
plt.title("CapHit vs ValueScore by Team")
plt.tight_layout()
plt.savefig("../outputs/graphs/caphit_vs_value.png")
plt.show()

# ========== PLOT 3: Correlation Heatmap ==========
plt.figure(figsize=(10, 6))
corr = df[["Goals", "Assists", "CapHit", "TOI", "PlusMinus", "GoalsPerMillion", "EfficiencyIndex", "ValueScore"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Player Stats")
plt.tight_layout()
plt.savefig("../outputs/graphs/player_feature_correlation_heatmap.png")
plt.show()
```
